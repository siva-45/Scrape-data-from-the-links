```python
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import pandas as pd

def search_query(query, num_results=2):
    # Perform a Google search and return a list of search results
    search_results = list(search(query, stop=num_results, pause=2))

    return search_results

def scrape_data_from_link(link):
    # Send an HTTP request to the link
    response = requests.get(link)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Get the title of the page
        title_tag = soup.title
        if title_tag:
            title = title_tag.text.strip()
            print("Title:", title)
        else:
            print("Title not found on the page:", link)
            title = None

        return title  # Return the title

        # You can expand this function to scrape more relevant data based on the query

    else:
        print("Failed to retrieve data from the website")

def save_to_csv(data, filename):
    # Convert the data to a pandas DataFrame
    df = pd.DataFrame(data)

    # Save the data to a CSV file
    df.to_csv(filename, index=False)
    print("Data saved successfully to", filename)

if __name__ == "__main__":
    queries = [
        "what industry in which Canoo operates",
        "size of canoo",
        "growth rate of Canoo",
        "current trends in that industry",
        "key players in  canoo",
        "Analyze Canoo's main competitors",
        "market share of canno",
        "what products or services is canno offereing",
        "pricing strategies of canoo",
        "marketing efforts of canoo",
        "Identify key trends in the market" ,
        "what is the recent  changes in consumer behavior related to this industry",
        "recent technological advancements in this industry",
        "Canoo's revenue, profit margins, return on investment, and expense structure"
    ]

    # Dictionary to store scraped data for each query
    scraped_data = {
        'Query': [],
        'Title': [],
        # Add more columns as needed for additional data
    }

    # Perform searches and scrape data for each query
    for query in queries:
        print("Searching for:", query)
        search_results = search_query(query)

        for i, link in enumerate(search_results, start=1):
            print(f"Scraping data from link {i}/{len(search_results)}")
            title = scrape_data_from_link(link)  # Get the title from the link
            if title:  # Check if title is not None
                # Store the query and title in the scraped_data dictionary
                scraped_data['Query'].append(query)
                scraped_data['Title'].append(title)
                # Add more columns as needed for additional data

    # Save the scraped data to a CSV file
    save_to_csv(scraped_data, 'scraped_data.csv')
```
